## Feature Selection

```python
numerical_df = top_manga_copy.select_dtypes(include = ['float','int'])
numerical_df.head()
```

```python
abs_corr_coeffs = top_manga_copy.corr()['Members'].abs().sort_values()
abs_corr_coeffs
```

```python
abs_corr_coeffs[abs_corr_coeffs > 0.14]
```

```python
# Drop columns with less than 0.14 correlation with Members
top_manga_copy = top_manga_copy.drop(abs_corr_coeffs[abs_corr_coeffs < 0.14].index, axis = 1)
```

```python
## Create a list of column names from documentation that are *meant* to be categorical
columns = ['Type', 'Status', 'Genres', 'Serialization' ]
```

```python
top_manga_copy.info()
```

```python
### How many unique values in each categorical column?
uniqueness_counts = top_manga_copy[columns].apply(lambda col: len(col.value_counts())).sort_values()
### Aribtrary cutoff of 10 unique values (worth experimenting)
# drop_nonuniq_cols = uniqueness_counts[uniqueness_counts > 10].index
# transform_df = top_manga_copy.drop(drop_nonuniq_cols, axis=1)
# drop_nonuniq_cols
# transform_df
uniqueness_counts
```

```python
text_cols = top_manga_copy.select_dtypes(include=['object', 'string'])

for col in columns:
    top_manga_copy[col] = top_manga_copy[col].astype('category')
    
# Create dummy columns and add back to the dataframe!
top_manga_copy = pd.concat([
    top_manga_copy, 
    pd.get_dummies(top_manga_copy.select_dtypes(include=['category']))
], axis=1).drop(text_cols,axis=1)
```

## Splitting the Data Into Train and Test Sets

```python
train = top_manga_copy.sample(frac = 0.8)
train
```

```python
#everything else that is not in the train set
test = top_manga_copy.loc[~top_manga_copy.index.isin(train.index)]
test
```

## Applying Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
model_columns = train.columns
#dropping these columns as this will be feeding the answer to the model
model_columns = model_columns.drop(['Favorites', 'Score Voted By', 'Members'])
```

```python
linear_model = LinearRegression()
linear_model.fit(train[model_columns], train['Members'])
linear_predictions = linear_model.predict(test[model_columns])
linear_mae = mean_absolute_error(test['Members'], linear_predictions)
linear_mae
```

```python
'''A reason for the high mae could be due a majority of the members being on the lower range
compared to the higher range, hence a bigger difference is seen when subtracted'''
train['Members'].sort_values().unique()
```

## Applying Decision Trees

```python
from sklearn.tree import DecisionTreeRegressor
tree_model = DecisionTreeRegressor(min_samples_leaf = 2) # min_samples_leaf = 2
tree_model.fit(train[model_columns], train['Members'])
tree_predictions = tree_model.predict(test[model_columns])
tree_mae = mean_absolute_error(test['Members'], tree_predictions)
tree_mae
```

## Applying Random Forests

```python
from sklearn.ensemble import RandomForestRegressor
forest_model = RandomForestRegressor(min_samples_leaf = 5)
forest_model.fit(train[model_columns],train['Members'])
forest_predictions = np.round(forest_model.predict(test[model_columns]))

forest_mae = mean_absolute_error(test['Members'], forest_predictions)
forest_mae
```

```python
test['predictions'] = forest_predictions
test[['Members','predictions']]
```

## Holdout Validation

Applying holdout validation as this is considered a more robust technique for testing a machine learning model's accuracy on new data the model wasn't trained on.

```python
train_top_manga = top_manga_copy
holdout_top_manga = top_manga_copy
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
# from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LogisticRegression
rf = RandomForestRegressor()
all_X = train_top_manga[model_columns]
all_y = train_top_manga['Members']
train_X, test_X, train_y, test_y = train_test_split(
    all_X, all_y, test_size=0.2,random_state=0)

rf.fit(train_X, train_y)
predictions= rf.predict(test_X)
scores = cross_val_score(rf, all_X, all_y, cv = 10, scoring = 'neg_mean_absolute_error')
accuracy = np.mean(scores)
print('score:', scores)
print('accuracy:', accuracy)
```

```python
# Testing on unseen data
holdout_predictions = rf.predict(holdout_top_manga[model_columns])

predicted_df = {'Members': holdout_top_manga['Members'],
                 'Predicted_Members': holdout_predictions}
predicted = pd.DataFrame(predicted_df)
predicted
```
