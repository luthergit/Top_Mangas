# Top Mangas

An online start-up company that focuses on mangas, comics, novels want to to enter the manga/comics/novels market and what to know what kind of manga would be the most popular. 

The aim of this project is to find the top best manga genres

The dataset used for this projects contains the top 500 ranked mangas (including novels, manhwas, manhuas, one-shots) from my [myanimelist](https://myanimelist.net) and is available [here](https://www.kaggle.com/ajpass/top-ranked-mangas-myanimelist-mal?select=top500mangaMAL.csv)

MyAnimeList(MAL) is an anime and manga social networking and social cataloging application website run by volunteers. The site provides its users with a list-like system to organize and score anime and manga.

## Top Popular Genres on MAL

From the dataset, the more popular the manga, the higher the number of members for that manga.

```python
popular_manga = top_manga_copy.sort_values('Members', ascending=False)
popular_manga
```

```python
popular_manga.plot(kind = 'line', x = 'Members', y = 'Popularity' ,figsize = (16,8))
plt.title('Popularity vs Members', fontsize=20)
plt.style.use('fivethirtyeight')
plt.legend('')
plt.xlim([140000,300000])
plt.ylabel('Popularity', x = -300)
plt.ylim([20, 0])
plt.tick_params(
    axis='x',           # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False)
```

```python
top_popular_manga = top_manga_copy.sort_values('Members', ascending=True)
```

```python
top_popular_genres = {}

for genres in top_popular_manga['Genres']:
    for genre in genres:
        if genre in top_popular_genres:
            top_popular_genres[genre] += 1
        else:
            top_popular_genres[genre] = 1
            
top_popular_genres_df = pd.DataFrame.from_dict(
    top_popular_genres, orient='index', columns=['count'])

top_popular_genres_df = top_popular_genres_df.sort_values('count').tail(20)
```

```python
top_popular_genres_df.plot(kind = 'barh', figsize = (16,8))
plt.title('Top 20 Popular Manga Genres on myanimelist', fontsize=20)
plt.style.use('fivethirtyeight')
plt.legend('')
```

From the graph, above Drama, Slice (of life) and comedy are still in the top three spots. However, **most popular genres have at least Drama in their genres.**

## Top Popular Running Genres vs. Top Finished Running Genres

```python
most_popular_finished = most_popular[most_popular['Status'] == 'Finished']
print (
    "There are", most_popular_finished.shape[0] ,
    "mangas that are finised")
```

```python
most_popular_running = most_popular[most_popular['Status'] == 'Publishing']
print (
    "There are", most_popular_running.shape[0] ,
    "mangas that are currently running")
```

```python
top_popular_running_genres = {}

for genres in most_popular_running['Genres']:
#     print (genres)
    for genre in genres:
#         print (genre)
        if genre in top_popular_running_genres:
            top_popular_running_genres[genre] += 1
        else:
            top_popular_running_genres[genre] = 1
            
top_popular_running_genres_df = pd.DataFrame.from_dict(
    top_popular_running_genres, orient='index', columns=['count'])

top_popular_running_genres_df = top_popular_running_genres_df.sort_values('count')
top_popular_running_genres_df.plot(kind = 'barh', figsize = (16,8))
plt.title('Genres for the Top Popular Running Manags on myanimelist', fontsize=20)
plt.style.use('fivethirtyeight')
plt.legend('')
```

```python
top_popular_finished_genres = {}

for genres in most_popular_finished['Genres']:
#     print (genres)
    for genre in genres:
#         print (genre)
        if genre in top_popular_finished_genres:
            top_popular_finished_genres[genre] += 1
        else:
            top_popular_finished_genres[genre] = 1
            
top_popular_finished_genres_df = pd.DataFrame.from_dict(
    top_popular_finished_genres, orient='index', columns=['count'])

top_popular_finished_genres_df = top_popular_finished_genres_df.sort_values('count')
top_popular_finished_genres_df.plot(kind = 'barh', figsize = (16,8))
plt.title('Genres for the Top Popular Finished Manags on myanimelist', fontsize=20)
plt.style.use('fivethirtyeight')
plt.legend('')
```

```python
top_20_popular_running_genres_df = top_popular_running_genres_df.tail(20)
top_20_popular_finished_genres_df = top_popular_finished_genres_df.tail(20)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)

fig.suptitle('Top 20 Running Genres vs. Top 20 Finished Genres', fontsize=30)

top_20_popular_running_genres_df.plot(kind = 'barh', figsize = (16,8), ax = ax1)
ax1.set_title('Top Running Genres', y = 0.99)
ax1.legend('')
plt.style.use('fivethirtyeight')
plt.legend('')


top_20_popular_finished_genres_df.plot(kind = 'barh', figsize = (16,8), ax = ax2)
ax2.set_title('Top Finshed Genres', y = 0.99)
ax2.legend('')
plt.style.use('fivethirtyeight')
```

For the top running genres, comedy, action and fantasy are the top three with more than 50 counts while for top finished genres, this is similar to the top popular (and ranked genres) with drama with over 200 counts with slice (of life) and comedy right after.

Almost 20 top popular finished mangas are drama followed by  slice (of life) and comedy. For the top 20 running mangas, the most occuring genres are comedy, action, fantasy with above 50 counts however, drama is in 6th place with a bit above 40 counts.

## Years Since Debut vs. Popularity for the Top 20 Running Mangas

```python
top_20_popular_running = most_popular_running.tail(20)
top_20_popular_finished = most_popular_finished.tail(20)
```

```python
top_20_popular_running['year_published'] = most_popular_running[
    'Published Dates'].copy().tail(20).str[10:14]
top_20_popular_running['years_since_debut'] = 2020 - top_20_popular_running[
    'year_published'].copy().astype('int64')

top_20_popular_finished['year_published'] = top_20_popular_finished[
    'Published Dates'].copy().tail(20).str[10:14]
top_20_popular_finished['years_since_debut'] = 2020 - top_20_popular_finished[
    'year_published'].copy().astype('int64')
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
figsize=(50, 100)

fig.suptitle('Years Since Debut vs. Popularity for the Top 20 Running Mangas', fontsize=50)

top_20_popular_running.plot(
    kind = 'barh', y = 'years_since_debut', x  = 'English Title', 
    figsize = (40,15), fontsize = 27, ax = ax1, xlabel = '', ylabel = "Years")
plt.style.use('fivethirtyeight')

most_popular_running.tail(20).plot(
    kind = 'barh', y = 'Members', x  = 'English Title', 
    figsize = (40,15), fontsize = 27, ax = ax2, xlabel = '', ylabel = 'Years')
plt.style.use('fivethirtyeight')

ax1.set_title('Years Since Debut', y = 1, fontsize = 30)
ax1.legend('')

ax2.set_title('Top Popular Mangas', y = 1, fontsize = 30)
ax2.legend('')
```

This is a very interesting revelation, the amount of years a manga has been out since it's debut has little to no correlation with it's popularity and all of these top 20 mangas all have anime adaptation except Yotsuba, Komi-san and Na Honjaman Level Up. The anime adaptation could be considered as an advertisement for the mangas helping it to reach more members. 

# Conclusion

Focusing on genres from the top 20 popular running mangas as it's top three **Attack on Titan, One Piece and Berserk** are the only running mangas to surpass 250,000 members. The genres will be top 12 popular running **comedy, action, fantasy, slice, shounen, drama, romance, adventure, school, sci-fi, seinen, supernatural** as a majority of popular running mangas have at least one or more of these genres. 

Therefore, I will recommend creating a manga that has genres all of which are from top 6 running genres - comedy, action, fantasy, slice, shounen, drama - to get better chance of becoming popular as members gravitate more to these genres. 

I will also recommend looking at the sales data on the top three running mangas, to analyse their demographic, besting selling region, advertising and spending. 

There is little to no correlation between the years since debut and the popularity and it appears that the main driving force for popularity are the genres. It maybe beneficial to create anime adaption of the manga series to enable it reach more members as this could be considered as an advertisement for the mangas helping it to reach more members as all the top 3 running mangas have anime adaptations.
